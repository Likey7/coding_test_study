```python

def get_rotate_key(M,key):
    rotate_key = [[0]*M for _ in range(M)]
    for i in range(M):
        for j in range(M):
            rotate_key[i][j] = key[M-1-j][i]
    
    return rotate_key

def solution(key, lock):
    answer = True
    M = len(key)
    N = len(lock)
    pad_size = N+2*(M-1)
    
    
    # lock 을 탐색하여 0인 (홈) 을 기록한다. 
    # 기록할 때 나중에 편하기 위해서 -> pad 된 상태의 인덱스를 확인 
    lock_zero = []
    lock_one  = []
    for i in range(N):
        for j in range(N):
            if lock[i][j] == 0 : 
                lock_zero.append((i+M-1,j+M-1))
            else :
                lock_one.append((i+M-1,j+M-1))
    
    #list comprehension 으로 2차원 배열 초기화 
    pad_lock = [[-1 for _ in range(pad_size)] for row in range(pad_size)] # M 의 테두리에 패딩을 단다,
    pad_together = [[-1 for _ in range(pad_size)] for row in range(pad_size)]
    
    # 기존 lock 패턴을 pad_lock 중심으로 옮겨준다. 
    for i in range(N):
        for j in range(N):
            pad_lock[i+M-1][j+M-1] = lock[i][j]
            pad_together[i+M-1][j+M-1] = lock[i][j]
            
    print(pad_lock)
    print(pad_together)
    # key 정방향인 것을 순서대로 더한다. 
    # key 1234 존재
    # 90 도 회전 
    key2 = get_rotate_key(M,key)
    key3 = get_rotate_key(M,key2)
    key4 = get_rotate_key(M,key3)
    
    # 전체 순회 
    # i,j 는 왼쪽 위 기준점 
    for i in range(N+M-1):
        for j in range(N+M-1):
            
            for test_key in [key, key2,key3,key4]:
                # pad together 초기화 
                
                pad_together = [[-1 for col in range(pad_size)] for row in range(pad_size)]
                for a in range(N):
                    for b in range(N):
                        pad_together[a+M-1][b+M-1] = lock[a][b]
                
                # together =  lock + key
                for k in range(M):
                    for l in range(M):
                        pad_together[i+k][j+l] = pad_together[i+k][j+l] +test_key[k][l]
                
                # i, j 기준으로 key 를 한번 꽂아본 상태
                # lock_zero 와 lock_one 의 인덱스가 전부 1이고 전부 1인지 확인 
                # 한번이라도 되면 참인 것 
                answer= True
                #하나라도 1이 아니면면 답이 아님 
                for a, b in lock_zero+lock_one:
                    if pad_together[a][b]!=1:
                        answer = False
                        break
                
                # lock_zero와 lock_one 을 모두 확인했는데도 flag == True 라면 return answer 
                if answer:
                    return answer
                    
        
    return answer
```

```python
# 입력 받는 부분
N = int(input())
K = int(input())
K_list = []
for _ in range(K):
    K_list.append(list(map(int,input().split())))
L = int(input())
time_list = []
turn_list = []
for _ in range(L):
    time,turn = input().split()
    time_list.append(int(time))
    turn_list.append(turn)

print(N)
print(K)
print(K_list)
print(L)
print(time_list)
print(turn_list)

############################################
    
# 2차원 배열 초기화
arr = [[1 for col in range(N+2)] for row in range(N+2)]

# 패딩 벽 0
for i in range(N+2):
    for j in range(N+2):
        if (i in [0,N+1] )or (j in [0,N+1]) : 
            arr[i][j]=0

# 사과 3 -> 
for (x,y) in K_list:
    arr[x][y] = 3
 

# 뱀의 몸이 차지하고 있는 곳  = 2   
arr[1][1] = 2

now = 0

# head, tail 을 리스로 만들면 수정했을 때 기존에 snake에 올린 리스트가 수정되어서 기록이 안됨
#head = [1,1]
head_x=1
head_y=1
#tail = [1,1]
tail_x=1
tail_y=1
side = 0 # 0 right, 90 up, 180 left, 270 down
is_apple = 0 # head 가 위치한 곳에 사과가 있으면 1 없으면 0
#length = 1

snake_queue = []
snake_queue.append([head_x,head_y])

timing = time_list[0]
time_list = time_list[1:]
timing_side = turn_list[0]
turn_list = turn_list[1:]

for row in arr:
    print(row)
    
###############################################

while True :
    # 시작하면서 시간 증가!
    now +=1
    print('now')
    print(now)
    
    # 방향에 따라 헤드 위치 이동 
    if side==0: 
        head_y = head_y+1
    elif side ==90:
        head_x = head_x-1
    elif side == 180 : 
        head_y = head_y-1
    elif side ==270 : 
        head_x = head_x+1
    
    print('head')
    print([head_x,head_y])
    
    # 스네이크 큐에 헤드 올림
    snake_queue.append([head_x,head_y])
    
    # head 가 이동한 위치가 벽인가? 몸인가? 확인 
    if arr[head_x][head_y]==0 or arr[head_x][head_y]==2:
        break
        
    # head 가 이동한 위치에 사과가 있는지 확인 
    if arr[head_x][head_y]==3: # 사과가 있으면 
        is_apple =1
        arr[head_x][head_y]=2 # 사과 먹음 -> 몸이 위에 있음
        # length+=1
        # 꼬리 위치 그대로 (snake 변동 없음 )
        
    else : # 사과가 없으면 
        is_apple = 0
        arr[head_x][head_y]=2 # 사과는 안먹었지만 땅을 먹음
        
        tail_x, tail_y= snake_queue[0][0], snake_queue[0][1] # 꼬리 리스트에서 지움
        snake_queue = snake_queue[1:]
        
        arr[tail_x][tail_y]=1 # 땅 원래대로 1로 변경 
        

    print('snake_queue')
    print(snake_queue)
    
    # 마지막에 초가 끝난 후 확인해야 할 것은 방향 전환 time 이 되었는가 이다
    if now == timing:
        # timing 이라면, 헤드 방향에 따라 전환
        if timing_side=='L':
            side = (side+90)%360            
        else :#timing_side =='D'
            side = (side+270)%360
            
           
        if len(time_list):
            timing = time_list[0]
            time_list = time_list[1:]
            timing_side = turn_list[0]
            turn_list = turn_list[1:]
        else:
            timing = -1
            timing_side = 0
    
    print('timing')
    print(timing)
    
    print('*********************************')

print(now)
```

```python

# 1. 바닥인가
# 2. 아래가 기둥인가
# 3. 왼쪽이 보인가
# + array 를벗어나지 않아야 함!
def col_check(arr, x, y):
    action=False
    if (y==0) :
        action=True
    elif (arr[x][y-1] ==0) : # y=!0 이니까 인덱스 벗어나지 않음
        action=True
    elif x>0 and (arr[x-1][y] ==1):
        action = True
    return action



# 1. 기둥 한족 끝/ 아래가 기둥
# 2. 왼쪽이 보 + 오른쪽이 보
# 3. 왼쪽이 보 + 오른쪽 아래가 기둥
def row_check(arr,x,y):
    
    action=False
    if y>0 and (arr[x][y-1]==0) :
        action=True
    elif (x>0 and x<(len(arr)-1)) and (arr[x-1][y]==1 and arr[x+1][y]==1) :
        action=True
    elif arr[x+1][y-1]==0 :
        action=True
    return action
    
    


def solution(n, build_frame):
    answer = []
    arr = [[-1 for col in range(n+1)] for row in range(n+1)]
    
    for frame in build_frame:
        print('frame : ',frame)
        # frame [x,y,a,b]
        x,y,a,b = frame[0], frame[1], frame[2], frame[3]
        flag=False
        
        
        if b==1: # 기둥/보 설치
            # 기둥 설치 가능할 때 
            if a==0:
                if col_check(arr,x,y):
                    flag=True
                    
            # 보 설치 가능할 때 
            else : # a==1
                if row_check(arr,x,y):
                    flag=True
                    
            if flag:
                print(x,y)
                print(arr)
                arr[x][y] = a
                answer.append([x,y,a])
            
            # 가능하지 않은 경우는 무시된다. 
        else: # 기둥 보 삭제 
            # 기둥 삭제
            temp = arr[x][y]
            arr[x][y] = -1
            
            if a ==0:
                # 왼쪽위 보 / 위 기둥 / 위 보  존재 가능한지 확인 
                if row_check(arr,x-1,y+1) and col_check(arr,x, y+1) and row_check(arr,x,y+1):
                    flag = True
                
            # 보 삭제 
            else: # a==1
                # 왼쪽 보 / 오른쪽 보  /오른쪽 기둥 
                if row_check(arr,x-1,y) and row_check(arr,x+1,y) and col_check(arr,x+1,y):
                    flag = True
                    
            if flag :
                del answer[answer.index([x,y,a])]
            else:
                arr[x][y]=temp
            
    answer.sort()
    return answer
```

```python

# 기둥을(0) 세울 수 있는가?
# 1. 바닥인가
# 2. 아래가 기둥인가
# 3. 왼쪽이 보인가
# 4. 지금 위치에 보가 있는가. 
# + array 를벗어나지 않아야 함!
def col_check(arr_col,arr_row, x, y):
    
    action=False
    if (y==0) :
        action=True
    elif (arr_col[x][y-1] ==0) : # y=!0 이니까 인덱스 벗어나지 않음
        action=True
    elif x>0 and (arr_row[x-1][y] ==1):
        action = True
    elif arr_row[x][y]==1:
        action = True
    return action


# 보를 (1) 설치할 수 있는가? 
# 1. 아래가 기둥
# 2. 왼쪽이 보 + 오른쪽이 보
# 3. 오른쪽 아래가 기둥
# 4. 
def row_check(arr_col,arr_row,x,y):
    action=False
    
    if y>0 and (arr_col[x][y-1]==0) :
        action=True
    elif (x>0 and x<(len(arr_row)-1)) and (arr_row[x-1][y]==1 and arr_row[x+1][y]==1) :
        action=True
    elif (y>0 and x<(len(arr_row)-1)) and arr_col[x+1][y-1]==0 :
        action=True
    return action
    
    


def solution(n, build_frame):
    answer = []
    arr_col = [[-1 for col in range(n+1)] for row in range(n+1)]
    arr_row = [[-1 for col in range(n+1)] for row in range(n+1)]
    
    for frame in build_frame:
        # frame [x,y,a,b]
        x,y,a,b = frame[0], frame[1], frame[2], frame[3]
        flag=False
        
        
        if b==1: # 기둥/보 설치
            # 기둥 설치 가능할 때 
            if a==0:
                if col_check(arr_col,arr_row,x,y):                    
                    arr_col[x][y] = a
                    answer.append([x,y,a])
                    
            # 보 설치 가능할 때 
            else : # a==1
                if row_check(arr_col,arr_row,x,y):                    
                    arr_row[x][y] = a
                    answer.append([x,y,a])
            
            # 가능하지 않은 경우는 무시된다. 
        else: # b==0 기둥 보 삭제             
            if a ==0:
                # 기둥 삭제
                temp = arr_col[x][y]
                arr_col[x][y] = -1
                # / 왼쪽위 보 / 위 기둥 / 위 보  존재 가능한지 확인 
                if row_check(arr_col,arr_row,x-1,y+1) and col_check(arr_col,arr_row,x, y+1) and row_check(arr_col,arr_row,x,y+1):
                    del answer[answer.index([x,y,a])]
                else:
                    arr_col[x][y]=temp
                
            # 보 삭제 
            else: # a==1
                
                temp = arr_row[x][y]
                arr_row[x][y] = -1
                # 왼쪽 보 / 오른쪽 보  /오른쪽 기둥 /현재우치 기둥
                if row_check(arr_col,arr_row,x-1,y) and row_check(arr_col,arr_row,x+1,y) and col_check(arr_col,arr_row,x+1,y) and col_check(arr_col,arr_row,x,y):
                    del answer[answer.index([x,y,a])]
                else:
                    arr_row[x][y]=temp
                    
            
    answer.sort()
    return answer
```
